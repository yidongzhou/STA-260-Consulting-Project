```python
def remove_non_ascii_1(text):
    return ''.join(i for i in text if ord(i)<128)
```

```python
class MyCorpus:
    """An iterator that yields sentences (lists of str)."""

    def __iter__(self):
        corpus_path = datapath('/Users/yihan/Downloads/test.txt')
        for line in open(corpus_path):
            # assume there's one document per line, tokens separated by whitespace
            yield utils.simple_preprocess(line)
```

```python
for i,j in enumerate(names):
    with open(names[i],encoding='utf8') as f:
        lines = f.readlines()
        tex = open('/Users/yihan/Downloads/test.txt','wt')
        for k in range(len(lines)):
            tex.write(remove_non_ascii_1(lines[k].replace('\t',' ')+'\n').lower())
    sentences = MyCorpus()
    model = gensim.models.Word2Vec(sentences=sentences,min_count=100,size=300)
    model.wv.save_word2vec_format(saves[i], binary=False)
```
